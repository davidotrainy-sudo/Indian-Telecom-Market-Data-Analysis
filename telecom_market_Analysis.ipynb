```python
import pdfplumber

def is_digital(pdf_path, threshold=100):
    """
    Returns True if the PDF contains enough selectable text
    to be treated as a digital PDF.
    """
    try:
        with pdfplumber.open(pdf_path) as pdf:

            total_text = ""

            for page in pdf.pages:

                text = page.extract_text()

                if text:
                    total_text += text

            return len(total_text.strip()) >= threshold

    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return False
```

```python
import os
import pdfplumber
import pandas as pd

# Folder containing PDFs
PDF_FOLDER = "drive/MyDrive/TRAI_PDFs"

# Keywords to identify the provider table
KEYWORDS = [
    "Reliance Jio",
    "Bharti Airtel",
    "Vodafone Idea",
    "Bharat Sanchar Nigam"
]

# Standardize column names
def standardize_columns(df):

    rename_dict = {
        "S.No.": "SNo",
        "S.N.": "SNo",
        "Name of the Service Provider": "Provider",
        "Service Provider": "Provider",
        "Subscriber base\n(In million)": "Subscribers"
    }

    df = df.rename(columns=rename_dict)

    return df


# Extract provider table from one PDF
def extract_provider_table(pdf_path):

    try:

        with pdfplumber.open(pdf_path) as pdf:

            for page in pdf.pages:

                tables = page.extract_tables()

                if not tables:
                    continue

                for table in tables:

                    table_text = str(table)

                    if any(k in table_text for k in KEYWORDS):

                        df = pd.DataFrame(
                            table[1:],
                            columns=table[0]
                        )

                        # Remove columns with None headers
                        df = df.loc[:, df.columns.notna()]

                        # Remove empty columns
                        df = df.dropna(axis=1, how="all")

                        # Standardize column names
                        df = standardize_columns(df)

                        # Keep only required columns
                        required = ["SNo", "Provider", "Subscribers"]

                        if not all(col in df.columns for col in required):
                            return None

                        df = df[required]

                        return df

    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")

    return None


# ----------------------------
# Main Program
# ----------------------------

master_df = pd.DataFrame()

processed = []
failed = []

files = os.listdir(PDF_FOLDER)

for file in files:

    pdf_path = os.path.join(PDF_FOLDER, file)

    df = extract_provider_table(pdf_path)

    if df is not None:

        df["Source_File"] = file

        master_df = pd.concat(
            [master_df, df],
            ignore_index=True
        )

        processed.append(file)

    else:

        failed.append(file)


print("="*60)
print("Processed PDFs :", len(processed))
print("Failed PDFs    :", len(failed))
print("="*60)

print("\nFirst few rows:")
print(master_df.head())

master_df.to_csv(
    "telecom_provider_data.csv",
    index=False
)

print("\nCSV saved as telecom_provider_data.csv")

print("\nFailed Files:")
for f in failed:
    print(f)
```

```python
import os
import re
import pdfplumber
import pandas as pd
from datetime import datetime


# Folder containing TRAI PDFs
folder = "drive/MyDrive/TRAI_PDFs"


all_data = []


# Function to extract date from PDF text
def extract_pdf_date(pdf_path):

    with pdfplumber.open(pdf_path) as pdf:

        for page in pdf.pages[:3]:   # check first 3 pages

            text = page.extract_text()

            if text:

                # Pattern: 31st December, 2025
                match = re.search(
                    r'(\d{1,2}(?:st|nd|rd|th)?\s+'
                    r'(January|February|March|April|May|June|July|August|September|October|November|December)'
                    r'[,\s]+\d{4})',
                    text,
                    re.IGNORECASE
                )

                if match:

                    date_text = match.group(1)

                    date_text = re.sub(
                        r'(st|nd|rd|th)',
                        '',
                        date_text
                    )

                    try:
                        return datetime.strptime(
                            date_text,
                            "%d %B, %Y"
                        )

                    except:

                        try:
                            return datetime.strptime(
                                date_text,
                                "%d %B %Y"
                            )

                        except:
                            pass

    return None



# Read every PDF
files = os.listdir(folder)


for file in files:

    if file.endswith(".pdf"):


        pdf_path = os.path.join(folder, file)

        print("Processing:", file)


        # Extract date from PDF
        pdf_date = extract_pdf_date(pdf_path)



        pdf = pdfplumber.open(pdf_path)


        provider_table = None



        # Find provider table
        for page in pdf.pages:

            tables = page.extract_tables()


            for table in tables:


                table_text = str(table).lower()


                if (
                    "jio" in table_text
                    and "airtel" in table_text
                    and "subscriber" in table_text
                ):

                    provider_table = table
                    break


            if provider_table:
                break



        # Extract providers
        if provider_table:


            for row in provider_table[1:6]:


                if len(row) > 1:


                    provider = row[1]


                    subscribers = next(
                        (
                            x for x in row[2:]
                            if x and x.strip()
                        ),
                        None
                    )


                    all_data.append(
                        {
                            "Provider": provider,
                            "Subscribers": subscribers,
                            "Date": pdf_date,
                            "Source_File": file
                        }
                    )


        else:

            print("Provider table not found:", file)



        pdf.close()



# Create dataframe

df = pd.DataFrame(all_data)



# Sort chronologically

df = df.sort_values(
    by="Date"
)



# Create readable date format

df["Date"] = df["Date"].dt.strftime(
    "%B %d, %Y"
)



# Save CSV

df.to_csv(
    "telecom_provider_data_chronological.csv",
    index=False
)



print("\nCompleted!")
print(df.head())

print(
    "\nSaved as telecom_provider_data_chronological.csv"
)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("telecom_market.db")

query = """
WITH market AS (

    SELECT
        Date,
        Provider,
        Subscribers,

        Subscribers * 100.0 /
        SUM(Subscribers) OVER(PARTITION BY Date)
        AS Market_Share

    FROM subscriber_data

),

growth AS (

    SELECT
        Provider,
        Date,
        Subscribers,

        (
        Subscribers -
        LAG(Subscribers)
        OVER(
            PARTITION BY Provider
            ORDER BY Date
        )
        )
        *100.0 /

        LAG(Subscribers)
        OVER(
            PARTITION BY Provider
            ORDER BY Date
        )

        AS Growth_Rate

    FROM subscriber_data

)

SELECT

m.Date,
m.Provider,
m.Subscribers,

ROUND(m.Market_Share,2)
AS Market_Share,

ROUND(g.Growth_Rate,2)
AS Growth_Rate,

ROUND(
SUM(m.Market_Share*m.Market_Share)
OVER(PARTITION BY m.Date),
2
)
AS HHI

FROM market m

LEFT JOIN growth g

ON m.Provider = g.Provider
AND m.Date = g.Date

ORDER BY
m.Date,
m.Market_Share DESC;
"""

result = pd.read_sql(query, conn)

result
```

```python
# Convert Date column
result["Date"] = pd.to_datetime(result["Date"])

# Create Day, Month, Year columns
result["Day"] = result["Date"].dt.day
result["Month"] = result["Date"].dt.month_name()
result["Year"] = result["Date"].dt.year

# Create readable date format
result["Date_Formatted"] = result["Date"].dt.strftime("%d-%B-%Y")

# Arrange columns
result = result[
    [
        "Date",
        "Date_Formatted",
        "Day",
        "Month",
        "Year",
        "Provider",
        "Subscribers",
        "Market_Share",
        "Growth_Rate",
        "HHI"
    ]
]

# Save CSV
result.to_csv(
    "TRAI_Telecom_Competition_TimeSeries_Final.csv",
    index=False
)

print("Downloaded file created successfully")
```
